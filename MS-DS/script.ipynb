```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs 
```

```python
X,y = make_blobs(1000,centers=4,random_state=42)
```

```python
X
```

```python
y
```

```python
def count(n):
    count=0
    for i in y:
        if i==n:
            count+=1
    print(count)

count(2)
count(0)
```
