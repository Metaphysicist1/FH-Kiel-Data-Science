```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vector = np.array([1, 2, 3, 4, 5])
```

```python
plt.plot(vector, marker='o')
plt.title('Line Plot of Vector')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
vectors = np.array([
    [1, 2],
    [3, 4],
    [-1, -2]
])

# Origin points for the vectors
origin = np.zeros(vectors.shape)

# Plot the vectors
colors = ['r', 'g', 'b']
plt.quiver(origin[:, 0], 7, vectors[:, 0], vectors[:, 1], angles='xy', scale_units='xy', scale=1, color=colors)
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Vector Arrows Visualization')
plt.grid()
plt.show()
```
