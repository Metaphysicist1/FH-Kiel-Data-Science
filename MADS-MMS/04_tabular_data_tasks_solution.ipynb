### Exercise 1.


```python
u = np.array([1,-2])
w = np.array([-1,2])
v = np.array([3,2])
vectors = np.array([u, w, v])
```

```python
origin = np.zeros(vectors.shape)
```

```python
colors = ['r', 'b', 'g']    
plt.quiver(origin[:,0],0, vectors[:,0], vectors[:,1], angles='xy', scale_units='xy', scale=1, color=colors)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid()
plt.show()
```

```python
uv = u+v
uw = u+w
u1_5 = u*1.5
u_v = u-v 
compute_V = 1.5*u + 0.5*v
```

```python
first_vectors = np.array([
    uv,uw,u1_5,u_v,compute_V
])

first_origin = np.zeros(first_vectors.shape)

colors = ['r','b','g','y','cyan']
plt.quiver(first_origin[:,0], 0, first_vectors[:,0], first_vectors[:,1], 
           angles='xy', scale_units='xy', scale=1, color=colors, 
           headwidth=5, headlength=7)

for i in range(len(first_vectors)):
    plt.text(first_vectors[i, 0], first_vectors[i, 1], f'V{i+1}', fontsize=12, color=colors[i])

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid(linestyle="--", linewidth=0.5)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([
    [1, 2, 3, 4.5],
    [3, 7, 2, 9],
    [5, 5, 2, 2]
])

# Take first two elements from each row for x and y components
vectors = A[:, :2]  # Extracting first two columns as x and y components

# Origins at (0,0)
origins = np.zeros_like(vectors)

# Colors for each vector
colors = ['r', 'g', 'b']

plt.figure(figsize=(6, 6))
plt.quiver(origins[:,0], origins[:,1], vectors[:,0], vectors[:,1], 
           angles='xy', scale_units='xy', scale=1, color=colors, headwidth=5, headlength=7)

# Label vectors
for i, vec in enumerate(vectors):
    plt.text(vec[0], vec[1], f'V{i+1}', fontsize=12, color=colors[i])

plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(linestyle="--", linewidth=0.5)
plt.title("Row Vectors as 2D Arrows")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(A.shape[1])  # Columns
y = np.arange(A.shape[0])  # Rows
X, Y = np.meshgrid(x, y)   # Create grid
Z = A                      # Heights from array

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k')

ax.set_xlabel('Columns')
ax.set_ylabel('Rows')
ax.set_zlabel('Values')
plt.title("3D Surface Plot of Matrix A")
plt.show()
```
