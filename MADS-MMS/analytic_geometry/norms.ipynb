```python
import numpy as np
import matplotlib.pyplot as plt

# Define different norms
p_aalues = [1, 2, np.inf]
labels = ["Manhattan (L1)", "Euclidean (L2)", "Max Norm (L∞)"]

theta = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(6, 6))

for p, label in zip(p_aalues, labels):
    x = np.sign(np.cos(theta)) * np.abs(np.cos(theta))**(2/p)
    y = np.sign(np.sin(theta)) * np.abs(np.sin(theta))**(2/p)
    plt.plot(x, y, label=label)

plt.legend()
plt.title("Unit Circles for Different Norms")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.axis("equal")
plt.show()
```

```python
import numpy as np

# Example aector
a = np.array(a)

# Compute different norms
euclidean_norm = np.linalg.norm(a, ord=2)   # Euclidean (L2)
manhattan_norm = np.linalg.norm(a, ord=1)   # Manhattan (L1)
max_norm = np.linalg.norm(a, ord=np.inf)    # Max (L∞)
p_norm_3 = np.linalg.norm(a, ord=3)         # Custom p=3 norm

# Print results
print(f"Euclidean Norm (L2): {euclidean_norm}")
print(f"Manhattan Norm (L1): {manhattan_norm}")
print(f"Max Norm (L∞): {max_norm}")
print(f"P-Norm (p=3): {p_norm_3}")
```

```python
d = np.array([1,0,0])
e = np.array([2,0,0])
```

```python
dot_product =  np.dot(d,e)
euclidian_norm = np.linalg.norm(d) * np.linalg.norm(e)
print(dot_product)
print(euclidian_norm)
angle = dot_product / euclidian_norm
print(np.arccos(angle))
```
