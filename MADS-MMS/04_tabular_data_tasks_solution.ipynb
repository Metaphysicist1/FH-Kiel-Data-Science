```python
import numpy as np
import matplotlib.pyplot as plt
```

### Exercise 1.


```python
u = np.array([1,-2])
w = np.array([-1,2])
v = np.array([3,2])
vectors = np.array([u, w, v])
```

```python
origin = np.zeros(vectors.shape)
```

```python
colors = ['r', 'b', 'g']    
plt.quiver(origin[:,0],0, vectors[:,0], vectors[:,1], angles='xy', scale_units='xy', scale=1, color=colors)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid()
plt.show()
```
